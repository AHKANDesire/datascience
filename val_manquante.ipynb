# Interpolation

```python
df = pd.DataFrame({'A':[1,2,np.nan,3,np.nan],'B':[1.2,7,3,0,8]})
df
```

```python
df['C'] = df['A'].interpolate()
df
```

```python
df['D'] = df.A.interpolate(method='spline', order=1)
df
```
