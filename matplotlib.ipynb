# exercice d'entrainement

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time=np.array([1,2,3,4,5,6,9])
concentration=np.array([5.5,7.2,11.8,13.6,19.1,21.7,29.4])
```

```python
fig,ax=plt.subplots()
ax.scatter(time,concentration, label="mesure")
ax.set_xlabel("Temps(h)")
ax.set_ylabel("concentrarion(mg/L)")
ax.set_title('concentration du produit en fonction du temps')

x=np.linspace(min(time), max(time) ,50)
y=2+3*x
ax.plot(x, y, color='green', ls="--", label="modèle")
ax.set_xlim(0, 10)
ax.set_ylim(0, 35)
ax.grid()
ax.legend(loc="upper left")
plt.show()
fig.savefig("mesure.png")
```

# Représentation en diagramme en bâtons
On souhaite maintenant représenter graphiquement la distribution des différentes bases dans une séquence d’ADN

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sequence = "ACGATCATAGCGAGCTACGTAGAA"
bases = ["A", "C", "G", "T"]
distribution = []
for i in bases:
    distribution.append(sequence.count(i))
distribution
x=np.arange(len(bases))

fig,ax=plt.subplots()
ax.bar(x, distribution)
ax.set_xticks(x,bases)
ax.set_ylim(0, 10)
ax.set_xlabel("bases")
ax.set_ylabel("nombres")
ax.set_title(f"Distribution des bases\n dans la séquence {sequence}")
plt.show()

```
